```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# นำเข้าข้อมูล

```python
BNN = pd.read_excel('Banana.xlsx', sheet_name = 'Property')
```

```python
BNN.index = BNN['ID']
BNN = BNN.drop(['ID', 'Storage type'],  axis = 1)
```

# สร้าง Model MLR

## จาก ตัวแปร L* a*  และ b* ทำนาย Brix

```python
X1 = BNN[['a*','b*']]
X2 = BNN[['a*','L*']]
X3 = BNN[['L*','b*']]
X4 = BNN[['L*','a*','b*']]
```

```python
A1 = BNN[['a*','b*']]
A2 = BNN[['a*','L*']]
A3 = BNN[['L*','b*']]
A4 = BNN[['L*','a*','b*']]
```

```python
Y = BNN[['acid']]
```

```python
B = BNN[['Firmness']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import metrics
```

```python
X2_train, X2_test, y_train, y_test = train_test_split(X2, Y, test_size=0.20, random_state=100)
MLR2 = LinearRegression()
MLR2.fit(X2_train, y_train)
```
