```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# นำเข้าข้อมูล

```python
BNN = pd.read_excel('Banana.xlsx', sheet_name = 'Property')
```

```python
BNN.index = BNN['ID']
BNN = BNN.drop(['ID', 'Storage type'],  axis = 1)
```

# สร้าง Model MLR

## จาก ตัวแปร L* a*  และ b* ทำนาย Brix

```python
X1 = BNN[['a*','b*']]
X2 = BNN[['a*','L*']]
X3 = BNN[['L*','b*']]
```

```python
A1 = BNN[['a*','b*']]
A2 = BNN[['a*','L*']]
A3 = BNN[['L*','b*']]
```

```python
Y = BNN[['acid']]
```

```python
B = BNN[['Firmness']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import metrics
```

```python
X1_train, X1_test, y_train, y_test = train_test_split(X1, Y, test_size=0.20, random_state=100)
MLR1 = LinearRegression()
MLR1.fit(X1_train, y_train)
```

```python
X2_train, X2_test, y_train, y_test = train_test_split(X2, Y, test_size=0.20, random_state=100)
MLR2 = LinearRegression()
MLR2.fit(X2_train, y_train)
```

```python
X3_train, X3_test, y_train, y_test = train_test_split(X3, Y, test_size=0.20, random_state=100)
MLR3 = LinearRegression()
MLR3.fit(X3_train, y_train)
```

```python
A1_train, A1_test, B_train, B_test = train_test_split(A1, B, test_size=0.20, random_state=100)
MLR4 = LinearRegression()
MLR4.fit(A1_train, B_train)
```

```python
A2_train, A2_test, B_train, B_test = train_test_split(A2, B, test_size=0.20, random_state=100)
MLR5 = LinearRegression()
MLR5.fit(A2_train, B_train)
```

```python
A3_train, A3_test, B_train, B_test = train_test_split(A3, B, test_size=0.20, random_state=100)
MLR6 = LinearRegression()
MLR6.fit(A3_train, B_train)
```

```python
MLR1.intercept_
MLR2.intercept_
MLR3.intercept_
MLR4.intercept_
MLR5.intercept_
MLR6.intercept_
```

```python
X1_train_prd = MLR1.predict(X1_train)
X2_train_prd = MLR2.predict(X2_train)
X3_train_prd = MLR3.predict(X3_train)
```

```python
A1_train_prd = MLR4.predict(A1_train)
A2_train_prd = MLR5.predict(A2_train)
A3_train_prd = MLR6.predict(A3_train)
```

```python
X1_test_prd = MLR1.predict(X1_test)
X2_test_prd = MLR2.predict(X3_test)
X3_test_prd = MLR3.predict(X3_test)
```

```python
A1_test_prd = MLR4.predict(A1_test)
A2_test_prd = MLR5.predict(A2_test)
A3_test_prd = MLR6.predict(A3_test)
```
